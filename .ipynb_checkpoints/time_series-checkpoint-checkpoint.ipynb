## Replicate Goyal & Welch (2008)

The Goyal & Welch Out-of-Sample (OOS) procedure is as follows:

1) Decide on an initial window of time, say $ s $ months (usually the first 20 years of data)

2) Estimate the following regression using data from $ t=0 $ to $ t=s $:

<font size='4'>

$
r^M_{t+1} = \alpha + \beta x_t + \epsilon_{t+1}
$

</font>


Where $ r^M_{t+1} $ is the equity risk premium in month $ t+1 $ (S&P500 total return minus the risk-free rate), and $ x_t $ is some predictor.

3) The above regression is estimated repeatedly over an expanding window starting with $ t=\{0,s\} $ to $ t=\{0,T\} $, where the sample-period increases by 1-month per iteration. Over each iteration, the estimated coefficients are stored (ie. $ \{\hat{\alpha_t}, \hat{\beta_t}\}^T_{t=s} $).

4) The stored coefficients are used to construct the predicted equity premium for month $ t+1 $ (as of month $ t $), based on the predictor in month $ t $ and the estimated coefficients from month $ t-1 $. That is: 

<font size='4'>

$
pred_t = \hat{\alpha}_{t-1} + \hat{\beta}_{t-1} x_t
$

</font>    
    
This is repeated to construct a full-time series of what is essentially a 1-step ahead forecast of the equity premium avoiding look-ahead bias.

5) The prediction is evaluated with OLS to see how well it performs, and whether it is significant:

<font size='4'>

$
r^M_{t+1} = a + b pred_t + e_{t+1}
$
    
</font>

```python
# Imports
import pandas as pd
import numpy as np
from datetime import datetime
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")
```

```python
# Functions

def annualize_return(monthly_return):
    '''
    Function to annualize the return of a single month
    '''    
    return (1 + monthly_return )**12 -1

def annualize_vol(monthly_vol):
    '''
    Function to annualize the standard devation of a single month
    '''    
    return monthly_vol * np.sqrt(12)
```

### Data
For sake of consistency with the 2008 paper, the monthly sample will be subsetted to start in 1927

```python
# Import data from Amit Goyal's site
dateparser = lambda x: datetime.strptime(x, '%Y%m')
data = pd.read_excel('goyal_data.xlsx', sheet_name='Monthly', header=0, index_col=0, parse_dates=True, date_parser=dateparser)
data = data.loc["1927-01-01":]
data.index = data.index + pd.offsets.MonthEnd(0)
```

```python
# Equity Premium
log_returns = np.log(1 + data['CRSP_SPvw'])
log_rf = np.log(1 + data['Rfree'])
data['EqPr'] = log_returns - log_rf
excess_returns = data['EqPr']
```

```python
# Annualized statistics
# From Goyal and Welcy 2008: "The equity premium had a mean (standard deviation) of ... 6.04% (19.17%) from 1927 to 2005"
mean = data['EqPr'].loc[:'2005'].describe()['mean']
vol = data['EqPr'].loc[:'2005'].describe()['std']

print(f'''
      The annualized average equity excess return is {annualize_return(mean): 0.2%}, 
      and the standard deviation is {annualize_vol(vol): 0.2%}
      ''')
```

### Empirical Replication

#### OOS Statistics

```python
def mse(y, y_hat):
    '''
    Computes MSE 
    '''
    errors = y - y_hat
    squared_errors = errors**2
    n = len(y)
    
    return (1/n) * squared_errors.sum()

def oos_r2(y, y_hat_model, y_hat_mean):
    '''
    Computes Out-of-Sample R-squared
    '''
    
    return 1 - (mse(y, y_hat_model) / mse(y, y_hat_mean))

def f_stat(y, y_hat_model, y_hat_mean):
    '''
    Computes MSE-F, the McCracken's (2004) F-statistic
    '''
    
    return (T - s - 1) * ((mse(y, y_hat_mean) - mse(y, y_hat_model)) / mse(y, y_hat_model))
```

#### In-sample

```python
reg_dp = sm.OLS(excess_returns[1:], 
                sm.add_constant(dp.shift(1).dropna())).fit()
reg_dp.summary()
```

#### Out-of-Sample

```python
dp_oos = sm.add_constant(dp.shift(1).dropna())
```

```python
# Step 1-3
s = 240 # first 20 years of data
coeffs = pd.DataFrame()

for t in range(0,len(excess_returns) - s):
    reg_dp_oos = sm.OLS(excess_returns[1:s+t+1], 
                    dp_oos.iloc[:s+t]).fit()
    coeffs.insert(0, t, reg_dp_oos.params)
    
coeffs = coeffs.T.iloc[::-1]
coeffs.index = dp.index[s:]
```

```python
# Step 4
pred = (dp_oos.iloc[240:] * coeffs.shift(1).dropna()).sum(axis=1)
pred.name = 'pred'+'_dp'
```

```python
# Step 5
reg_dp_oos_2 = sm.OLS(excess_returns[241:], 
                sm.add_constant(pred)).fit()
```

```python
ex_ret_hat = reg_dp_oos_2.predict()
```

```python
# Mean Model
ex_ret_mean_hat = excess_returns.iloc[240:] - excess_returns.expanding(min_periods=240).mean().iloc[240:]
```

```python
oos_r2(excess_returns[241:], ex_ret_hat, ex_ret_mean_hat.shift(1).dropna())
```
